# Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface
for drawing attractive and informative statistical graphic luded

- Provides a layer of abstraction hence simpler to use
- better aesthetics
- more graphs included

# Main Classification
- Relational Plot
- Distribution Plot
- Categorical Plot
- Regression Plot
- Matrix Plot
- Multiplots

```python
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips
```

## sns.scatterplot(data, x, y, hue, style, size)

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
```

## sns.histplot(data, x, bins, hue, element)

```python
sns.histplot(data=tips, x='total_bill', bins=10, hue='sex', element='step')
```

## sns.displot(data, x, kind)

```python
sns.displot(data=tips, x='total_bill', kind='hist')
```

**Bivariate Histogram**
A bivariate histogram bins the data within rectangles that tile the plot
and then shows the count of observations within each rectangle with the fill color

```python
# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data=tips, x='total_bill', y='tip', kind='hist')
```

## sns.kdeplot(data, x, hue, fill)

Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel,
producing a continuous density estimate

```python
sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True)
```

## Boxplot
A boxplot is a standardized way of displaying the distribution of data based on a five number
summary (â€œminimumâ€, first quartile [Q1], median, third quartile [Q3] and â€œmaximumâ€). It can
tell you about your outliers and what their values are. Boxplots can also tell you if your data is
symmetrical, how tightly your data is grouped and if and how your data is skewed.

## sns.boxplot(data, x, y, hue)

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
```

## sns.barplot(data, x, y, hue, estimator)

```python
sns.barplot(data=tips, x='sex', y='total_bill', hue='smoker')
```

## sns.countplot(data, x, hue)

```python
sns.countplot(data=tips, x='sex', hue='day')
```
